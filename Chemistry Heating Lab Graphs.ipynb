```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.optimize import curve_fit

time = np.arange(0, 660 , 30)
mgcl = np.array([22.5, 23., 23.5, 24., 24., 24., 24., 23.75, 23.5, 23.5, 23.5, 23.5, 23.25, 23., 23., 23., 23.])
cacl = np.array([30., 33., 34., 34., 34., 34., 33.5, 33., 33., 33., 33., 33., 33., 32.5, 32., 32., 32., 32., 31.5])
acetate = np.array([18., 18., 17., 17., 17., 17., 17., 17., 17., 17., 17., 17., 17.5, 17.5, 17.5, 17.5, 17.5, 17.5, 17.5, 17.5, 17.5, 17.5])
naoh = np.array([30., 31., 41., 44., 44., 43., 42., 41., 40., 39.])
nhcl = np.array([15., 15., 15., 15., 15., 15., 15.5, 15.5, 15.5, 15.5, 15., 15., 16., 16., 15.5])

sns.regplot(time[:len(mgcl)], mgcl)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature ($^\circ$Celsius)")
plt.title("Time Plot of Temperature of Magnesium Chloride Solution")
plt.show()

sns.regplot(time[:len(cacl)], cacl)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature ($^\circ$Celsius)")
plt.title("Time Plot of Temperature of Calcium Chloride Solution")
plt.show()

sns.regplot(time, acetate)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature ($^\circ$Celsius)")
plt.title("Time Plot of Temperature of Sodium Acetate Solution")
plt.show()

sns.regplot(time[:len(naoh)], naoh)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature ($^\circ$Celsius)")
plt.title("Time Plot of Temperature of Sodium Hydroxide Solution")
plt.show()

sns.regplot(time[:len(nhcl)], nhcl)
plt.xlabel("Time (Seconds)")
plt.ylabel("Temperature ($^\circ$Celsius)")
plt.title("Time Plot of Temperature of Ammonium Chloride Solution")
plt.show()
```
